Question 1b


Create a list showing each playerâ€™s first and last names as well as the total salary they earned in the major leagues.

    Answer 1b: see vandy_player_salaries dataframe

````python
people = pd.read_sql("SELECT * FROM people", con=engine)
people.info()
````

````python
p=people[["playerid","namefirst","namelast"]]
p
````

````python
vandy_info=p['playerid'].isin(vandy_unique.playerid)
vandy_idfl=p.loc[vandy_info]
vandy_idfl
````

````python
vandy_idfl.info()
````

````python
########I'm at the point where I need to pull in salaries df, then keep the playerid and salary as a new df, then merge that new df with vandy_idfl on playerid.
### But be careful, look at salaries table for a couple of players and see if they need a groupby on playerid first before merge because of diff mlb teams and pay per year?
````

````python
salaries = pd.read_sql("SELECT * FROM salaries;", con=engine)
salaries.info()
````

````python
s=salaries[["playerid", "salary"]]
s
````

````python
# vandy_info=p['playerid'].isin(vandy_unique.playerid)
# vandy_idfl=p.loc[vandy_info]
# vandy_idfl
salary_info=s['playerid'].isin(vandy_idfl.playerid)
salaries=s.loc[salary_info]
salaries
````

````python
# https://www.codegrepper.com/code-examples/python/pandas+python+group+by+for+one+column+and+sum+another+column
s_players=salaries.groupby(['playerid'], as_index=False)['salary'].sum()
vandy_salaries=s_players.sort_values('salary', ascending=False)
````

````python
vps=vandy_salaries.merge(vandy_idfl, on='playerid')
vps
````

````python
vps['full_name']=vps['namefirst']+ str(' ') + vps['namelast']
vps
````

````python
vandy_player_salaries=vps[["full_name","salary"]]
vandy_player_salaries
````
